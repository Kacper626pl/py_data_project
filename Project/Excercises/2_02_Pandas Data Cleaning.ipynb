```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## ðŸŸ© Remove Rows with Missing Values (2.2.1)

```python
df_cleaned = df
```

```python
df_cleaned = df_cleaned.dropna(subset='salary_year_avg')
df_cleaned
```

---

## ðŸŸ¨ Remove Duplicate Rows (2.2.2)

```python
df_cleaned = df_cleaned.drop_duplicates()
```

---

## ðŸŸ¨ Fill Missing Values in Column (2.2.3)

```python
df_filled = df_cleaned['job_skills'].fillna("Unknown",inplace=True)
```
